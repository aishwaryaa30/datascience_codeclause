```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
iris =  pd.read_csv('Iris.csv')
iris
```

```python
# How many datapoints are there?
iris.shape
```

```python
# What are the features of the data?
iris.columns
```

```python
# How many flowers of each species?
iris['Species'].value_counts()
```

```python
# Plotting a 1-D Scatter Plot
iris_setosa = iris.loc[iris['Species'] == 'Iris-setosa']
iris_virginica = iris.loc[iris['Species'] == 'Iris-virginica']
iris_versicolor = iris.loc[iris['Species'] == 'Iris-versicolor']

plt.plot(iris_setosa['SepalLengthCm'], np.zeros_like(iris_setosa['SepalLengthCm']), 'o')
plt.plot(iris_virginica['SepalLengthCm'], np.zeros_like(iris_virginica['SepalLengthCm']), 'o')
plt.plot(iris_versicolor['SepalLengthCm'], np.zeros_like(iris_versicolor['SepalLengthCm']), 'o')
plt.grid()
plt.show()
```

```python
# Plotting a 2-D graph
iris.plot(kind='scatter', x='SepalLengthCm', y='SepalWidthCm')
plt.show()
```

```python
sns.set_style('whitegrid')
sns.pairplot(iris, hue='Species', height=3).add_legend()
plt.show()
```

```python
sns.set_style('whitegrid')
sns.FacetGrid(iris, hue='Species', height=4).map(plt.scatter, 'PetalLengthCm', 'PetalWidthCm').add_legend()
plt.show()
```
